<h1> Importing Libraries & Loading Initial Info </h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = sns.load_dataset("titanic")

print(df.head())
print(df.info())
print(df.describe())
print(df.isna().sum())
```

<h1>Survival Count (Countplot) </h1>

```python
plt.figure()
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

The visualization shows that a majority of passengers did not survive the Titanic disaster. The overall survival rate was considerably lower than the death rate, indicating the severity of the tragedy.

<h1>Survival by Gender (Countplot)  </h1>

```python
plt.figure()
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
```

The graph illustrates that females had a significantly higher survival rate compared to males. This suggests that gender played an important role in rescue priority during the evacuation.

<h1> Age Distributuon (Histogram) </h1>

```python
plt.figure()
sns.histplot(df['age'], bins=20, kde=True, color = 'black')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Most passengers were between 20 and 40 years old, forming the largest age group onboard. There were comparatively fewer children and elderly passengers in the dataset.

<h1> Survival By Passenger Class (Grouped / Aggregation) </h1>

```python
survival_by_class = df.groupby('pclass')['survived'].mean()

survival_by_class.plot(kind='bar')
plt.title("Average Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

The plot shows that 1st class passengers had the highest survival rate, while 3rd class passengers had the lowest. This indicates that passenger class strongly influenced survival chances.

<h1> Fare Distribution By Class (Histogram) </h1>

```python
plt.figure()
sns.boxplot(x='pclass', y='fare', data=df, color = 'green')
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

The visualization demonstrates that 1st class passengers paid much higher fares than those in lower classes. Additionally, fare variation is wider in higher passenger classes.

<h1> Age VS Fare (Scatter Plot) </h1>

```python
plt.figure()
plt.scatter(df['age'], df['fare'], color = 'red')
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

The scatter plot indicates no strong relationship between age and fare. Instead, fare differences appear to be more influenced by passenger class than by age.
